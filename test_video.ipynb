```python
import cv2
from IPython.display import display, clear_output
import matplotlib.pyplot as plt

# 打开视频文件
cap = cv2.VideoCapture("/mnt/data_hdd1/zby/Seattle_50/4_S_Jackson_NS_11_12_7_50.mp4")

# 获取视频的总帧数
total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))

# 设置从第 100 帧开始播放
start_frame = 150
cap.set(cv2.CAP_PROP_POS_FRAMES, start_frame)
# 创建一个 matplotlib 图像
plt.figure(figsize=(6, 6))

frame_counter = 0  # 帧计数器
frame_counter = start_frame
update_frequency = 50  # 每 100 帧更新一次显示

while True:
    ret, frame = cap.read()
    if not ret:
        break

    frame_counter += 1

    # 每隔一定帧数更新显示
    if frame_counter % update_frequency == 0:
        # 显示每一帧
        frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        plt.imshow(frame_rgb)
        plt.axis('off')  # 不显示坐标轴
        print(frame_counter)

        # 更新显示
        clear_output(wait=True)
        display(plt.gcf())
        #break

cap.release()
```
